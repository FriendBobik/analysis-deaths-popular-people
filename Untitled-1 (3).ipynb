```python
import os
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np

all_months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

root_folder = "deaths"
os.makedirs(root_folder, exist_ok=True)

def get_entries_from_page(url):
    headers = {"User-Agent": "MyWikipediaBot/1.0 (myemail@example.com)"}
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"{url} {response.status_code})")
        return pd.DataFrame(columns=["Name", "Age", "Info"])
    
    soup = BeautifulSoup(response.text, "html.parser")
    content_div = soup.find("div", class_="mw-parser-output")
    if not content_div:
        return pd.DataFrame(columns=["Name", "Age", "Info"])
    
    list_items = content_div.find_all("li")
    entries = []
    
    for li in list_items:
        text = li.get_text().strip()
        
        # Пропускаем строки, содержащие символ "^"
        # Чтобы не тригерилось на ссылки
        if "^" in text:
            continue
        
        # Удаляем ссылки вида [число] в конце строки
        clean_text = re.sub(r"\[\d+\]", "", text).strip()
        # Разбиваем строку на 3 части: имя, возраст и остальное
        parts = clean_text.split(",", 2)
        if len(parts) == 3:
            name = parts[0].strip()
            age = parts[1].strip()
            info = parts[2].strip()
            if info.endswith('.'):
                info = info[:-1].strip()
            entries.append((name, age, info))
    
    # Убираем дубликаты
    unique_entries = list({entry for entry in entries})
    df = pd.DataFrame(unique_entries, columns=["Name", "Age", "Info"])
    df = df.sort_values(by="Name")
    return df

# for year in range(1986, 2026):
#     if year == 2025:
#         months = all_months[:2]
#     else:
#         months = all_months

#     year_folder = os.path.join(root_folder, str(year))
#     os.makedirs(year_folder, exist_ok=True)
    
#     for month in months:
#         url = f"https://en.wikipedia.org/wiki/Deaths_in_{month}_{year}"
#         df = get_entries_from_page(url)
#         csv_filename = os.path.join(year_folder, f"{month}.csv")
#         df.to_csv(csv_filename, index=False, encoding="utf-8")

# Убираем -
def fix_age(age):
    if pd.isnull(age):
        return age
    age_str = str(age)
    if "–" in age_str:
        return age_str.split("–")[0].strip()
    elif "-" in age_str:
        return age_str.split("-")[0].strip()
    else:
        return age_str.strip()

# for dirpath, dirnames, filenames in os.walk(root_folder):
#     for filename in filenames:
#         if filename.endswith(".csv"):
#             file_path = os.path.join(dirpath, filename)
#             try:
#                 df = pd.read_csv(file_path, on_bad_lines='skip', engine='python')
#             except Exception as e:
#                 print(f"Ошибка чтения файла {file_path}: {e}")
#                 continue
        
#             df["Age"] = df["Age"].apply(fix_age)
#             df.to_csv(file_path, index=False, encoding="utf-8")
```

```python
import os
import pandas as pd

groupings = {
    "Suicide": [
        "suicide", "self-harm", "self inflicted", "self-inflicted", 
        "committed suicide", "took his own life", "took her own life", 
        "ended his life", "ended her life", "self-destruction", "self-killed"
    ],
    "Overdose": [
        "overdose"
    ],
    "Kill": [
        "shot", "gunshot", "bullet", "firearm", "gunfire", "bullet wound", 
        "shooting", "gunshot wound", "firearm injury", "gun violence", "beating", "explosion",
        "american airlines flight 11", "united airlines flight 175" ,"united airlines flight 93", "american airlines flight 77","september 11 attacks"
    ],
    "Cancer & Malignancies": [
        "cancer", "malignancy", "leukemia", "lymphoma", "tumor", "carcinoma", 
        "neoplasm", "sarcoma"
    ],
    "Cardiovascular Diseases": [
        "heart", "cardiac arrest", "heart attack", "myocardial infarction", 
        "arrhythmia", "myocardial", "aneurysm", "aortic dissection", 
        "cardiomyopathy", "myocarditis", "ventricular fibrillation", "sudden death", 
        "cardiac rupture", "cardiac tamponade", "coronary", "ischemic heart disease"
    ],
    "Neurological Conditions": [
        "stroke", "cva", "brain", "intracerebral hemorrhage", "cerebral edema", 
        "encephalitis", "brain injury", "hemorrhagic stroke", "ischemic stroke", 
        "subdural hematoma", "Alzheimer"
    ],
    "Respiratory Disorders": [
        "pneumonia", "respiratory", "respiratory failure", "copd", 
        "pulmonary embolism", "pulmonary edema", "ards", "acute respiratory distress", 
        "bronchitis", "asthma"
    ],
    "Hepatic & Renal Failure": [
        "liver", "cirrhosis", "hepatic failure", "liver failure", 
        "renal failure", "kidney", "end-stage renal disease"
    ],
    "External Causes & Injuries": [
        "fall", "accident", "crash", "drowning", "trauma", "choking", 
        "electrocution", "burns", "fracture", "asphyxiation", "collision", "blunt force"
    ],
    "Infections & Complications": [
        "complications", "infection", "sepsis", "septicemia", "meningitis", 
        "pancreatitis", "shock", "hemorrhage", "bleeding", "bacterial infection", 
        "viral infection", "fungal infection", "endotoxin", "inflammatory", "covid"
    ]
}

group_counts = {group: 0 for group in groupings.keys()}
group_counts["Other"] = 0

for dirpath, _, filenames in os.walk(root_folder):
    for filename in filenames:
        if filename.endswith(".csv"):
            filepath = os.path.join(dirpath, filename)
            try:
                df = pd.read_csv(filepath)
            except Exception as e:
                print(f"Ошибка при чтении файла {filepath}: {e}")
                continue

            if "Info" not in df.columns:
                continue

            for info in df["Info"]:
                if pd.isna(info):
                    continue
                text = str(info).lower()
                matched_groups = set()
            
                for group, keywords in groupings.items():
                    for kw in keywords:
                        if kw in text:
                            matched_groups.add(group)
                            break

                if not matched_groups:
                    group_counts["Other"] += 1
                else:
                
                    for group in matched_groups:
                        group_counts[group] += 1

sorted_group_counts = sorted(group_counts.items(), key=lambda x: x[1], reverse=True)

for group, cnt in sorted_group_counts:
    print(f"{group}: {cnt}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

all_categories = list(groupings.keys()) + ["Other"]
data_by_age = {}


for dirpath, _, filenames in os.walk(root_folder):
    for filename in filenames:
        if filename.endswith(".csv"):
            filepath = os.path.join(dirpath, filename)
            try:
                df = pd.read_csv(filepath)
            except Exception as e:
                print(f"Ошибка при чтении {filepath}: {e}")
                continue

            if "Age" not in df.columns or "Info" not in df.columns:
                continue

            df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

            for _, row in df.iterrows():
                age = row["Age"]
                if pd.isna(age):
                    continue
                age = int(age)
                info = row["Info"]
                if pd.isna(info):
                    continue
                text = str(info).lower()
                matched_groups = set()
                for group, keywords in groupings.items():
                    for kw in keywords:
                        if kw in text:
                            matched_groups.add(group)
                            break
                if age not in data_by_age:
                    data_by_age[age] = {cat: 0 for cat in all_categories}
                if not matched_groups:
                    data_by_age[age]["Other"] += 1
                else:
                    for group in matched_groups:
                        data_by_age[age][group] += 1

all_ages = sorted(data_by_age.keys())
rows = []
for age in all_ages:
    row = {"Age": age}
    for cat in all_categories:
        row[cat] = data_by_age[age].get(cat, 0)
    rows.append(row)

df_age = pd.DataFrame(rows)
df_age = df_age.sort_values("Age")


```

```python

data_by_age_music = {}

music_keywords = [
    "music", "song", "melody", "harmony", "rhythm", "beat", "tune",
    "chord", "lyrics", "composition", "symphony", "orchestra", "band",
    "concert", "album", "genre", "instrument", "guitar", "piano",
    "drums", "violin", "singer", "vocal", "choir", "opera", "rock",
    "pop", "jazz", "blues", "classical", "hip-hop", "rap", "EDM",
    "electronic", "folk", "metal", "punk", "reggae", "country",
    "bass", "treble", "note", "scale", "tempo", "acoustic", "live",
    "recording", "mixing", "mastering", "DJ", "producer", "festival",
    "soundtrack", "musical", "karaoke", "anthem", "ballad", "cover",
    "remix", "single", "EP", "LP", "vinyl", "cassette", "streaming",
    "playlist", "headphones", "speakers", "microphone", "studio"
]



for dirpath, _, filenames in os.walk(root_folder):
    for filename in filenames:
        if filename.endswith(".csv"):
            filepath = os.path.join(dirpath, filename)
            try:
                df = pd.read_csv(filepath)
            except Exception as e:
                print(f"Ошибка при чтении {filepath}: {e}")
                continue

            if "Age" not in df.columns or "Info" not in df.columns:
                continue

            df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

            for _, row in df.iterrows():
                age = row["Age"]
                if pd.isna(age):
                    continue
                age = int(age)
                info = row["Info"]
                if pd.isna(info):
                    continue
                text = str(info).lower()

                if not any(kw in text for kw in music_keywords):
                    continue

                matched_groups = set()
                for group, keywords in groupings.items():
                    for kw in keywords:
                        if kw.lower() in text:
                            matched_groups.add(group)
                            break

                if age not in data_by_age_music:
                    data_by_age_music[age] = {cat: 0 for cat in all_categories}
                    data_by_age_music[age]["Total"] = 0

                if not matched_groups:
                    data_by_age_music[age]["Other"] += 1
                else:
                    for group in matched_groups:
                        data_by_age_music[age][group] += 1

                data_by_age_music[age]["Total"] += 1

rows = []
for age in sorted(data_by_age_music.keys()):
    row = {"Age": age}
    for cat in all_categories:
        row[cat] = data_by_age_music[age].get(cat, 0)
    row["Total"] = data_by_age_music[age]["Total"]
    rows.append(row)

df_music_categories = pd.DataFrame(rows)
df_music_categories = df_music_categories.sort_values("Age")

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Категории
categories = [
    "Cancer & Malignancies", "Cardiovascular Diseases", "Infections & Complications",
    "External Causes & Injuries", "Neurological Conditions", "Respiratory Disorders",
    "Hepatic & Renal Failure", "Kill", "Suicide", "Overdose"
]

percent_total = [31.93, 20.68, 15.11, 7.72, 6.83, 5.76, 4.12, 3.94, 3.33, 0.59]
percent_musicians = [32.95, 18.38, 12.33, 5.52, 7.20, 5.12, 12.23, 2.31, 2.90, 1.07]

x = np.arange(len(categories))
width = 0.35  

fig, ax = plt.subplots(figsize=(12, 8))
bars_total = ax.bar(x - width/2, percent_total, width, label='Total')
bars_musicians = ax.bar(x + width/2, percent_musicians, width, label='Музыканты')

ax.set_ylabel('Процент (%)')
ax.set_title('Сравнение процентов заболеваний/смертей: Total vs. Музыканты')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

for i in range(len(categories)):
    total_val = percent_total[i]
    musician_val = percent_musicians[i]
    ratio = musician_val / total_val if total_val else 0
    max_height = max(total_val, musician_val)-0.75
    ax.text(x[i], max_height + 1, f"{ratio:.2f}x", ha="center", va="bottom", fontsize=10)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("3.png")

plt.show()
```
